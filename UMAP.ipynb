```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
import umap
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/MNoichl/UMAP-examples-mammoth/refs/heads/master/mammoth_a.csv')

#Subsample of N elements
N = 3000
X = df.sample(N).values

#plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

ax.scatter(X[:,0],X[:,1],X[:,2], c= X[:,2])
ax.set_title('Mammoth')
fig.show()
```

```python
#To rescale data
X_scaled = StandardScaler().fit_transform(X)

#2d embedding with UMAP
n_neighbors = 10

reducer = umap.UMAP(n_neighbors = n_neighbors)
embedding = reducer.fit_transform(X_scaled)
```

```python
#Plot of the results
plt.scatter(embedding[:,0],embedding[:,1], c = X[:,2])
```
